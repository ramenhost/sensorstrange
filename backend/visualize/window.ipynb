```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sen_df = pd.read_csv('../data/sensor_log01.log.bak')
init_time = sen_df.at[0,'timestamp']
sen_df['timestamp'] -= init_time
sen_df = sen_df.set_index('timestamp')
sen_df.head()
```

```python
acc_df = sen_df.loc[sen_df['sensor'].str.contains('Acce')]
acc_df.plot(subplots=True)
```

```python
gyro_df = sen_df.loc[sen_df['sensor'].str.contains('Gyro')]
gyro_df.plot(subplots=True)
```

```python
key_df = pd.read_csv('../data/keypress_log01.log.bak')
key_df['timestamp'] -= init_time
key_df = key_df.set_index('timestamp')
key_df.head()
```

```python
legend = {'H':'red', 'E':'green', 'L':'blue', 'O':'yellow', 'N':'violet'}
```

```python
acc_df.loc[acc_df.index<5000].plot()
for index, row in key_df.loc[key_df.index<5000].iterrows():
    if(row['action'] == 'ACTION_DOWN'):
        plt.axvspan(index-100, index+100, color=legend[row['key']], alpha=0.25)
```

```python
acc_df.loc[(acc_df.index>18500) & (acc_df.index<20500)].plot()
for index, row in key_df.loc[(key_df.index>18500) & (key_df.index<20500)].iterrows():
    if(row['action'] == 'ACTION_DOWN'):
        plt.axvspan(index-100, index+100, color=legend[row['key']], alpha=0.25)
```
